````python
num = input("Enter a number sequence to check if it is a palindrome :")

def is_palindrome_int(n):
    num = int (n)
    original, rev = num, 0
    while( num > 0):
        rev = rev *10 + num % 10 
        num//= 10
    return original == rev

def is_palindrome_string(n):
    s = str(n)
    return s == s[::-1]
        
print( "is palindrome" if is_palindrome_int(num) and is_palindrome_string(num) else " not a palindrome")

````

### 4. A three digit number is said to be an “Armstrong number” if the sum of the third power of its individual digits is equal to the number itself.Write a program to check whether a number is armstrong or not.

````python
def is_armstrong_bool(num):
    n = len(str(num))
    temp = num
    total = 0
    while temp > 0:
        digit = temp % 10
        total += digit ** n
        temp //= 10
    return total == num

def is_armstrong_str(num):
    dgts = str(num)
    n = len(dgts)
    return int(num) == sum(int(d)**n for d in dgts)

num = input("Enter a number to check if Armstrong: ")
n = int(num)

print("is armstrong" if is_armstrong_bool(n) and is_armstrong_str(num) else "not an armstrong")
````

If exact change is not possible, then display **-1**.

````python
def get_payable(available_1s, available_5s, amount):
    required_5s = min(amount // 5, available_5s)
    remainder = amount - (required_5s * 5)

    if remainder <= available_1s:
        return [required_5s, remainder]
    else:
        return[-1,-1]

available_1s = int(input("Enter number of 1's coin available :"))
available_5s = int(input("Enter number of 5's notes available :"))
amount = int(input("Enter amount payable :"))

payable = get_payable(available_1s, available_5s, amount)

if payable[0] != -1:
    print("Number of 5's notes required:", payable[0])
    print("Number of 1's coins required:", payable[1])
else:
    print(-1)

    
````

### 11. Write a Python program to generate the next 15 leap years starting from a given year. Populate the leap years into a list and display the list.

````python
def get_15_leap_years(starting_year):
    years = []
    year = starting_year
    while len(years) < 15:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            years.append(year)
        year += 1
    return years

start_year = int(input("Enter a year: "))
print("The next 15 leap years are:", get_15_leap_years(start_year))
````

- Consider each number from **num1** to **num2** (num2 inclusive). Populate the number into a list, if the below conditions are satisfied:  
  - Sum of the digits of the number is a multiple of 3  
  - Number has only two digits  
  - Number is a multiple of 5  
- Display the maximum element from the list  
- In case of any invalid data or if the list is empty, display **-1**


````python
def max_num_thing(n1, n2):
    num_list = []
    if n2 > n1:
        for n in range(n1,n2+1):
            if n >= 10 and n <=99 and sum(int(num) for num in str(n)) % 3 == 0 and n % 5 == 0:
                num_list.append(n)
    else:
        return -1

    return max(num_list)
            
        
n1 = int(input("Enter num1: "))
n2 = int(input("Enter num2: "))

print('The max is: ', max_num_thing(n1,n2))
````

**Note:**

### 18. The road transport corporation (RTC) of a city wants to know whether a particular bus-route is running on profit or loss.

Assume that the following information is given:

- **Price per litre of fuel** = 70  
- **Mileage of the bus** in km/litre of fuel = 10  
- **Price (Rs) per ticket** = 80  

The bus runs on multiple routes having different distances in kms and number of passengers.  

Write a function to calculate and return the **profit earned (Rs)** in each route.  
Return **-1** in case of loss.


````python
fuel_price = 70
mileage = 10
ticket_price = 80

def get_PL(distances, passengers):
    pl_list = []
    for d,p in zip(distances,passengers):
        cost = (d / mileage) * fuel_price
        revenue = p * ticket_price
        profit = revenue -cost
        pl_list.append(profit if profit > 0 else -1)

    return pl_list

distances = [int(km) for km in input("Enter distances for individual routes: ").split()]
passengers = [int(km) for km in input("Enter passengers per routes: ").split()]

print('The profit / loss list is: ', get_PL(distances, passengers))
````

### 19. Given a string containing uppercase characters (A–Z), compress the string using **Run Length Encoding**.  
Repetition of a character has to be replaced by storing the length of that run.

Write a Python function which performs the run length encoding for a given string and returns the run length encoded string.

Provide different string values and test your program.


````python
def run_length_encoding(string):
    encoded = ''
    i = 0
    length = len(string)

    while i < length:
        count = 1
        while i + 1 < length and string[i] == string[i+1]:
            count += 1
            i += 1
        encoded += str(count) + string[i]
        i += 1

    return encoded

string = input("Enter uppercase characters (A-Z) string to compress: ")
print("The compressed string is:", run_length_encoding(string))
````
